```python
#Importing the necessary libraries

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
url = 'https://www.rozee.pk/'
```

```python
driver.get(url)
```

```python
# Wait for the search bar to be clickable

search_bar = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.ID, 'search'))
)

```

```python
#searching for AI?ML Engineer jobs

search_bar.send_keys('AI/ML Engineer' + Keys.RETURN)
```

```python
bar = driver.find_elements(By.XPATH, '/html/body/div[3]/div/div[2]/div[2]')

titles = driver.find_elements(By.XPATH, '//div[@class="jobt float-left"]//h3[@class="s-18"]')
companys = driver.find_elements(By.CLASS_NAME, 'cname')
```

```python
#extracting titles og jobs
rozee_titles = []
for title in titles:
    rozee_titles.append(title.text)
#     print(title.text)
rozee_titles
```

```python

# Find the element job link
job_link_element = driver.find_elements(By.XPATH, "//h3[@class='s-18']/a")
rozee_links=[]

# Extract the href attribute of the job link

for i in job_link_element:
    rozee_links.append(i.get_attribute('href'))
    
rozee_links
```

```python
#extracting company names
import re
rozee_compny =[]
for company in companys:
    if company:
        company_name = re.split(',|\n ', company.text)[0].strip()
        rozee_compny.append(company_name)
rozee_compny
```

```python
# extracting experience level

rozee_exp= []
experience_level_elements = driver.find_elements(By.CSS_SELECTOR, 'span.func-area-drn')

for element in experience_level_elements:
    rozee_exp.append(element.text)
#     print(element.text)
rozee_exp
```

```python
#location
rozee_loc =[]
for company in companys:
    if company:
        locations = re.split(',|\n ', company.text)[1].strip()
        rozee_loc.append(locations)
rozee_loc
```

```python
gender=[]
min_education=[]

from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


# Find all jobs
j = driver.find_elements(By.XPATH, '//*[@id="jobs"]')
job_title=driver.find_elements(By.XPATH, '//h3[@class="s-18"]')

# Iterate through each title
for i in j:
    for title in job_title:

        # Scroll to the title and click on it
        driver.execute_script("arguments[0].scrollIntoView(true);", title)
        i.click()
        
        #Gender
        # Wait for the gender details to be visible
        gender_details = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.XPATH, '//*[@id="job-content"]/div[3]/div/div[8]/div[2]'))
        )
        
        # Scroll to the gender details and get the text
        driver.execute_script("arguments[0].scrollIntoView(true);", gender_details)
        genders = gender_details.text
         # Append the gender details to the list
        gender.append(genders)
        
      
        #Minimum Experience
        experience_details = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.XPATH, '//*[@id="job-content"]/div[3]/div/div[10]/div[2]'))
        )
        
        # Scroll to the gender details and get the text
        driver.execute_script("arguments[0].scrollIntoView(true);", experience_details)
        exp = experience_details.text
         # Append the gender details to the list
        min_education.append(exp)
        

        
        # Wait for 2 seconds before proceeding to the next product
        time.sleep(2)
        driver.execute_script("window.scrollTo(0, 500);")
        
```

```python
#creating dataframe
rozee_dataset=pd.DataFrame({'Job_Title': rozee_titles,'Company':rozee_compny, 'Location': rozee_loc,'Experience': rozee_exp
                         , 'Gender':gender, 'Min_Eductaion':min_education, "Job_Links":rozee_links })
rozee_dataset
```

```python
indeed_bar = driver.find_elements(By.XPATH, '//*[@id="mosaic-provider-jobcards"]')

indeed_titles = driver.find_elements(By.XPATH, '//*[@id="mosaic-provider-jobcards"]/ul/li[1]')

indeed_companys = driver.find_elements(By.CLASS_NAME, 'jobTitle.css-1u6tfqq.eu4oa1w0')
```

```python

# Extract the text of each job title element
job_titles = []
for job_title_element in indeed_companys:
    job_titles.append(job_title_element.text)

job_titles    
```

```python

#company names
company_names =[]
company_name_element = driver.find_elements(By.XPATH, "//span[@data-testid='company-name' and contains(@class, 'css-1x7z1ps') and contains(@class, 'eu4oa1w0')]")
# company_name = company_name_element.text

for i in company_name_element:
    company_names.append(i.text)
    
company_names
```

```python
#job links indeed
job_links = driver.find_elements(By.CLASS_NAME, 'jcs-JobTitle')
indeed_links =[]
# Iterate through the links 
for link in job_links:
    href = link.get_attribute('href')
    indeed_links.append(link.get_attribute('href'))

```

```python
indeed_links
```

```python

location_elements = driver.find_elements(By.XPATH, "//*[@id='mosaic-provider-jobcards']/ul/li/div/div[1]/div/div[1]/div/table[1]/tbody/tr/td/div[2]/div/div")

# Extract locations
indeed_location=[]
for i, location_element in enumerate(location_elements, 1):
    indeed_location.append(location_element.text)
indeed_location    
```
