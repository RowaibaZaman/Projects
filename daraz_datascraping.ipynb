```python
#Importing libraries

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
url = 'https://www.daraz.pk/'
```

```python
driver.get(url)
```

```python
search_bar = driver.find_element(By.XPATH, '/html/body/div[1]/div/div/div[1]/div/div/div[2]/div/div[2]/form/div/div[1]/input[1]')
search_bar.send_keys('Mobile phones' + Keys.RETURN)
```

```python
ad_grid = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[3]/div/div/div[1]')

ad_cards = ad_grid.find_elements(By.CLASS_NAME, 'gridItem--Yd0sa')

prices = ad_grid.find_elements(By.CLASS_NAME, 'price--NVB62')

```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException

prod_title = []
prod_rating = []  
brand = [] 
reviews = []  
specification = []  
price = []
shipping_status=[] 
seller_rating=[]   # Sellerrating
sot=[] 
daraz_varification = []
dilvery_status = []

for r in range(1, 6):
    for i in range(len(ad_cards)):
        try:
            product_grid = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[3]/div/div/div[1]/div[2]')
            p = product_grid.find_elements(By.CLASS_NAME, 'gridItem--Yd0sa')[i]
            driver.execute_script("arguments[0].scrollIntoView(true);", p)
            p.click()

        #     PRICE       
            product_price = driver.find_element(By.XPATH, '//*[@id="module_product_price_1"]/div/div/span')
            price.append(product_price.text)
            time.sleep(2)

            # TITLE       
            product_title = driver.find_element(By.CLASS_NAME, 'pdp-mod-product-badge-title')
            prod_title.append(product_title.text)
            time.sleep(2)

            # BRAND      
            brand_ = driver.find_element(By.XPATH, '//*[@id="module_product_brand_1"]/div/a[1]')
            brand.append(brand_.text)
            time.sleep(3)

            #shipping status
            try:
                shipping=driver.find_element(By.XPATH, '//*[@id="module_seller_delivery"]/div/div/div[3]/div/div[1]/div/div[1]/div[1]/div[1]/span[1]')
                shipping_status.append(shipping.text)
                time.sleep(2)
            except:
                shipping_status.append("Not Specified")

            #varification
            try:
                varification=driver.find_element(By.XPATH, '//*[@id="module_product_title_1"]/div/div/img')
                daraz_varification.append("Daraz Mall Verified")
                time.sleep(2)  
            except:
                daraz_varification.append("Not specified")

            #dilvery status  
            try:
                dilvery_s=driver.find_element(By.XPATH, '//*[@id="module_seller_delivery"]/div/div/div[3]/div/div[1]/div/div[1]/div[1]/div[1]/span[1]')
                dilvery_status.append(dilvery_s.text)
                time.sleep(2) 
            except:
                 dilvery_status.append("Not Specified")


            driver.execute_script("window.scrollTo(0, 500);")

            # RATING
            wait = WebDriverWait(driver, 5)  # adjust the timeout as needed
            try:
                product_rating = wait.until(EC.presence_of_element_located((By.XPATH, '//span[@class="score"]')))
                prod_rating.append(product_rating.text)
            except TimeoutException:
                prod_rating.append("No Rating")
            time.sleep(1)

            # REVIEWS 
            wait = WebDriverWait(driver, 2)
            try:
                review = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'review-content-sl')))
                reviews.append(review.text)
            except TimeoutException:
                reviews.append("No Review")
            time.sleep(1)

            # SPECIFICATION
            try:
                spec = driver.find_element(By.XPATH, '//*[@id="module_product_detail"]/div/div')
                specification.append(spec.text)
            except NoSuchElementException:
                specification.append("No Specification")
            time.sleep(3)   

            #seller rating
            sr= driver.find_element(By.XPATH, '//*[@id="module_seller_info"]/div/div[2]/div[1]/div[2]')
            seller_rating.append(sr.text)
            time.sleep(1)


            score= driver.find_element(By.XPATH, '//*[@id="module_seller_info"]/div/div[2]/div[2]/div[2]')
            sot.append(score.text)
            time.sleep(3)

            driver.back()
            time.sleep(2)

        except NoSuchElementException as e:
            print(f"Element not found for ad_cards[{i}]. Skipping...")


        
    move = f'(//*[@id="root"]/div/div[3]/div/div/div[1]/div[3]/div[2]/div/ul/li[{r+2}])'
    frwd_button = driver.find_element(By.XPATH, move)
    frwd_button.click()
    time.sleep(2)  
```

```python

import pandas as pd

daraz_data =pd.DataFrame({'Product Title': prod_title, 'Product Rating': prod_rating,
                         'Price': price, 'Brand':brand,"Shipment Status":shipping_status,"Seller Rating":seller_rating,
                          "Ship on Time Score":sot, "Mall Varification":daraz_varification,
                          "Delivery Status":dilvery_status, "Reviews":reviews, 'Specification': specification 
                         })
daraz_data
```

```python
daraz_data.to_csv("Daraz_dataset.csv")
```
